# Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Importing Dataset and Preprocessing

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

# Training on Linear Regression

```python
## linear regression
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

# Training on Polynomial Regression

```python
## polynomial regression
from sklearn.preprocessing import PolynomialFeatures
## degree means for example: in eq y = b0 + b1x1 + b1x1^2 (till square power)
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
linear_reg_2 = LinearRegression()
linear_reg_2.fit(X_poly, y)
```

# Plotting Linear Regression Results

```python
plt.scatter(X, y, color="red")
plt.plot(X, lin_reg.predict(X), color="blue")
plt.title("Truth or Bluff (Linear Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

# Plotting Polynomial Regression

```python
plt.scatter(X, y, color="red")
plt.plot(X, linear_reg_2.predict(X_poly), color="blue")
plt.title("Truth or Bluff (Polynomial Linear Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

# Smoothening the edges to make it look like curves

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color="red")
plt.plot(X_grid, linear_reg_2.predict(poly_reg.fit_transform(X_grid)), color="blue")
plt.title("Truth or Bluff (Polynomial Linear Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

# Testing On Linear Regression

```python
## first [] means rows and second [] is columns
print(lin_reg.predict([[6.5]]))
```
